```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Visualization on Amazon Best Books Selling Dataset

```python
df=pd.read_csv("Amazon_BestSelling_Books_500.csv")
```

```python
df.head()
```

# Dataset Overview

This Dataset contains information about Amazon best selling books including ratings,prices and categories.

```python
df.info()
```

# Dataset Information

The dataset contains categorical and numerical columns used for visualization and analysis.

```python
df.hist(figsize=(10,8))
plt.show()
```

# Histogram Observation 
Histogram shows the distribution of numerical columns in the dataset.

```python
sns.countplot(x='Category',data=df)
plt.xticks(rotation=90)
plt.show()
```

## Category Analysis

Some categories appear more frequently than others.

```python
sns.boxplot(x='Category',y='Price (USD)',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(x='Price (USD)',y='Rating',data=df)
plt.xticks(rotation=90)
plt.show()
```

## Relationship Analysis

Scatter plot shows the relationship between price and rating.

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True),annot=True)
plt.show()
```

## Correlation Analysis

Heatmap displays correlation between numerical variables.

```python
df['Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```

## Category Percentage

Pie chart represents percentage distribution of categories.

```python
sns.lineplot(x='Reviews',y='Rating',data=df)
plt.show()
```

# Conclusion

Data visualization helped identify trends,patterns and relationships within the dataset using graphical analysis.
